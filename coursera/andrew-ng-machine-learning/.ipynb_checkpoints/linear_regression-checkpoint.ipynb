```python
"""
Generate random data for linear regression and plot it.
TODO: Try to find best fit line
"""

from matplotlib import pyplot as plt
import numpy as np

np.random.seed(5)
X = np.arange(100)
y = 3 * X + 40 + np.random.normal(0, 60, 100)

intersects, slopes, costs = [], [], []
for intersect in np.arange(30, 50):
    for slope in np.arange(2.5, 3.5, 0.01):
        cost = 0
        for xi, yi in zip(X, y):
            cost += ((slope * xi) - yi) ** 2
        cost = cost / (2 * len(X))

        intersects.append(intersect)
        slopes.append(slope)
        costs.append(cost)

# print("cost: {}, slope: {}".format(params[0], params[1]))

# plt.plot(X, y, "o")
# plt.plot([0, 100], [20, 320], color='k', linestyle='-', linewidth=1)
# plt.plot([0, 100], [25, 325], color='g', linestyle='-', linewidth=1)
plt.plot(slopes, costs)
plt.xlabel("Slope")
plt.ylabel("Cost")
plt.xlim(2.5, 3.5)
plt.show()

# import plotly as py
# import plotly.graph_objs as go
# py.tools.set_credentials_file(username='KondratiukViacheslav', api_key='W3TDxJV8CaM0mMmLjEjE')

# trace = go.Scatter(
#     x = slopes,
#     y = costs
# )

# data = [trace]

# py.plotly.iplot(data, file_id='line-mode')
```
